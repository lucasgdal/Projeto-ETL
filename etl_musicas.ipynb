```python
import pandas as pd
```

```python
#Importando o dataset

musicas = pd.read_csv('musicas.csv', encoding="ISO-8859-1")
```

```python
#Panorama geral do data frame

display(musicas)
```

```python
#Verificando todas as colunas
colunas_musicas = musicas.columns
for colunas in colunas_musicas:
    print(colunas)
```

```python
#analise de valores nulos
musicas.isnull()
```

```python
# Inserindo novos dados nos valores nulos
musicas['Artist'] = musicas['Artist'].fillna('desconhecido')
```

```python
# Removendo os valores nulos
musicas.dropna(axis=1, inplace=True)
```

```python
# Renomeando colunas
musicas.rename(columns={
    'Track': 'Músicas',
    'Album Name' : 'Álbuns',
    'Artist' : 'Artistas',
    'Release Date' : 'Data de Lançamento',
    'All Time Rank' : 'Rank'
}, inplace=True)
```
